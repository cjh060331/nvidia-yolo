```python
from ultralytics import YOLO
import cv2
import torch
from google.colab.patches import cv2_imshow
```

```python
# YOLO v11 모델 다운로드 및 불러오기
model = YOLO('yolo11x.pt')  # YOLO v11 모델 경로 (업데이트 필요할 수 있음)
```

```python

# 3. 동영상 업로드
from google.colab import files
uploaded = files.upload()

video_path = list(uploaded.keys())[0]  # 업로드한 첫 번째 파일 사용
```

```python


# 4. 동영상 읽기
cap = cv2.VideoCapture(video_path)

# 동영상 저장을 위한 설정
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('output.mp4', fourcc, cap.get(cv2.CAP_PROP_FPS),
                        (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # YOLO v11을 이용한 객체 탐지
    results = model(frame)

    # 탐지된 객체 그리기
    for result in results:
        for box in result.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            label = result.names[box.cls[0].item()]
            conf = box.conf[0].item()

            # 바운딩 박스 그리기
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, f'{label} {conf:.2f}', (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # 결과 프레임 저장
    out.write(frame)

cap.release()
out.release()
cv2.destroyAllWindows()

# 동영상 재생을 위해 파일을 Colab에 표시
from IPython.display import display, HTML

display(HTML("""

"""))

print("분석 완료! 'output.mp4' 파일이 생성되었습니다.")
```
